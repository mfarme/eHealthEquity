# Data source
Data from FCC manually downloaded from [FCC National Broadband Map](https://broadbandmap.fcc.gov/data-download/)


Closest data availability by county was June 30, 2022.

```python
df = pd.read_csv('./bdc_us_broadband_summary_by_geography_J22_03nov2023.csv', dtype={'geography_id': str})
```

```python
df.head()
```

```python
#filter to only include counties 
df = df[df['geography_type'] == 'County'].copy()
df.head()
```

# Variable definition 
t1_2_3_7_s3_r = 
Calculated percentage of units for broadband serviceable
locations contained within the geography for which
providers report residential fixed broadband service with
Copper, Cable, Fiber to the Premises, Unlicensed Fixed
Wireless, or Licensed Fixed Wireless technology and
speeds of at least 25 / 3 Mbps

```python
col_keep = [
    'geography_desc_full',
    't1_2_3_7_s3_r'
]

df = df[col_keep].copy()
df.head()
```

```python
#seperate out the state and county
df['state'] = df['geography_desc_full'].str.split(',').str[1].str.strip()
df['county'] = df['geography_desc_full'].str.split(',').str[0].str.strip()
df.drop(columns=['geography_desc_full'], inplace=True)
df.head()
```

```python
#export to csv
df.to_csv('./broadband_structural_jun_22.csv', index=False)
```
